## average rating by date

```python
data['Day']= data['Timestamp'].dt.date
data_avg=data.groupby(['Day']).count()
plt.figure(figsize=(25,3))
plt.plot(data_avg.index,data_avg['Rating'])
```

## Averge rating by week

```python
data['Week']= data['Timestamp'].dt.strftime('%y-%U')
data_weekavg=data.groupby(['Week']).count()
plt.figure(figsize=(25,3))
plt.plot(data_weekavg.index,data_weekavg['Rating'])
```

## average by Month

```python
data['Month']= data['Timestamp'].dt.strftime('%m-%U')
data_monthavg=data.groupby(['Month']).count()
plt.figure(figsize=(25,3))
plt.plot(data_monthavg.index,data_monthavg['Rating'])
```

## Average rating by month by course

```python
data['Month']= data['Timestamp'].dt.strftime('%Y-%m')
data_monthavg_crs=data.groupby(['Month','Course Name']).mean().unstack()
data_monthavg_crs.plot(figsize=(25,10))
```

## What day are people happiest

```python
data['Weekdays']=data["Timestamp"].dt.strftime('%A')
data['Daynumber']=data['Timestamp'].dt.strftime('%w')
weekday_average=data.groupby(['Weekdays','Daynumber']).mean()
weekday_average=weekday_average.sort_values('Daynumber')
plt.figure(figsize=(15,3))
plt.plot(weekday_average.index.get_level_values(0),weekday_average['Rating'])
```
